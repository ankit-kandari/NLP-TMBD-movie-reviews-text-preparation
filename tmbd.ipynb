```python
def sh_conversion(text):
    new_text = []
    for w in text.split():
        if w in short:
            new_text.append(short[w])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
#remove stopwords

from nltk.corpus import stopwords
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def remove_stop(text):
    l1 = []
    doc = nlp(text)
    for i in doc:
        if not i.is_stop:
            l1.append(i.text)
    return " ".join(l1)
```

```python
df1["review"]=df1["review"].apply(remove_stop)
```

```python
##lemmetization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wl = WordNetLemmatizer()
```

```python
def lemmatize_text(text):
    lemmatized_words = [wl.lemmatize(word) for word in text]
    return ' '.join(lemmatized_words)
```

```python
df1["review"]=df1["review"].apply(lemmatize_text)
```
